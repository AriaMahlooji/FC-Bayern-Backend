```python
def resultInThisMinute(minute, matchNumber):
    homeTeamGoalCounter=0
    awayTeamGoalCounter=0
    finalMatchStatus = matchDetailsSub2013_2021.iloc[matchNumber]['Final Match Status']
    if(matchDetailsSub2013_2021.iloc[matchNumber]['HomeTeamScoresMinutes'] != 'null'):
        for i in range(len(matchDetailsSub2013_2021.iloc[matchNumber]['HomeTeamScoresMinutes'])):
            if int(matchDetailsSub2013_2021.iloc[matchNumber]['HomeTeamScoresMinutes'][i]) <= minute :
                homeTeamGoalCounter+=1
    if(matchDetailsSub2013_2021.iloc[matchNumber]['AwayTeamScoresMinutes'] != 'null'):
        for i in range(len(matchDetailsSub2013_2021.iloc[matchNumber]['AwayTeamScoresMinutes'])):
            if int(matchDetailsSub2013_2021.iloc[matchNumber]['AwayTeamScoresMinutes'][i]) <= minute :
                awayTeamGoalCounter+=1
    if(matchDetailsSub2013_2021.iloc[matchNumber]['HomeOrAway']=='Home'):
        if(homeTeamGoalCounter > awayTeamGoalCounter):
            return 'Won' + ' to ' + finalMatchStatus
        if(homeTeamGoalCounter < awayTeamGoalCounter):
            return 'Lost'+ ' to ' + finalMatchStatus
        if(homeTeamGoalCounter == awayTeamGoalCounter):
            return 'Draw'+ ' to ' + finalMatchStatus
    
    if(matchDetailsSub2013_2021.iloc[matchNumber]['HomeOrAway']=='Away'):
        if(homeTeamGoalCounter > awayTeamGoalCounter):
            return 'Lost' + ' to ' + finalMatchStatus
        if(homeTeamGoalCounter < awayTeamGoalCounter):
            return 'Won' + ' to ' + finalMatchStatus
        if(homeTeamGoalCounter == awayTeamGoalCounter):
            return 'Draw' + ' to ' + finalMatchStatus
        
    
```

```python
def resultInThisMinuteInNumber(minute, matchNumber):
    homeTeamGoalCounter=0
    awayTeamGoalCounter=0
    finalMatchStatus = matchDetailsSub2013_2021.iloc[matchNumber]['Final Match Status']
    if(matchDetailsSub2013_2021.iloc[matchNumber]['HomeTeamScoresMinutes'] != 'null'):
        for i in range(len(matchDetailsSub2013_2021.iloc[matchNumber]['HomeTeamScoresMinutes'])):
            if int(matchDetailsSub2013_2021.iloc[matchNumber]['HomeTeamScoresMinutes'][i]) <= int(minute) :
                homeTeamGoalCounter+=1
    if(matchDetailsSub2013_2021.iloc[matchNumber]['AwayTeamScoresMinutes'] != 'null'):
        for i in range(len(matchDetailsSub2013_2021.iloc[matchNumber]['AwayTeamScoresMinutes'])):
            if int(matchDetailsSub2013_2021.iloc[matchNumber]['AwayTeamScoresMinutes'][i]) <= int(minute) :
                awayTeamGoalCounter+=1
    
    return str(homeTeamGoalCounter) + ' – ' + str(awayTeamGoalCounter)
    
```
